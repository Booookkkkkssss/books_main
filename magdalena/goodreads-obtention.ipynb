```python

def get_book_url():
    
    '''
    this function takes in a variable that identifies 
    HTML sections of a Goodreads search webpage and 
    returns a list of the urls for books with the word 
    'by' in the title
    '''

    # empty list
    book_ids = []

    for p in range(1, 101):
        
        # url to go to next page
        page_changer = f'https://www.goodreads.com/search?page={p}&q=by&qid=0msFdG9v9K&search%5Bfield%5D=title&search_type=books&tab=books&utf8=%E2%9C%93'
    
        # getting the response
        response = requests.get(page_changer)
        
        print(f'Page {p} status code : {response.status_code}')
        
        # creating soup obj
        soup = BeautifulSoup(response.content, 'html.parser')

        # url of book title
        book_url = soup.find_all('a', class_ = 'bookTitle')



        # looping through book_url variable
        for i in range(0, len(book_url)):


            # accessing the exact part of the output
            href = book_url[i]['href']

            print(href)

            # appending to list & cleaning up url
            book_ids.append(re.findall(r'\d.*', href))

        time.sleep(7)

        print(f'Page {p} complete.')

    return book_ids
```

```python
df = pd.read_csv('2000_book_ids.txt', header = None, names = ['book_url_seg'])

base = 'https://www.goodreads.com/book/show/'

collect = pd.DataFrame(columns = ['title','summary','year_published','author','review_count',
                                  'number_of_ratings','length','genre','rating','reviews'])

counter = 0

for i, item in enumerate(df['book_url_seg']):
    
    url = base + item
    
    # create initial web request
    response = requests.get(url)
    
    # convert to soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    print(response.status_code)
    print(counter)
    
    #--------title-------------------------------------------
    
     # save to Variable
    title_tag = soup.find('h1', attrs = {'class': 'Text Text__title1'}).text
    
    # looking at the h1 tag aria.
    # title = title_tag['aria-label'].split(': ')[1]
    print(f'Book Title: {title}')
    
    #----author-----------------------------------------------
    
    # find where author is stored
    soup.find('span', class_ = "ContributorLink__name")
    # save the text to a variable
    auth = soup.find('span', class_ = "ContributorLink__name").text
    print(f'Book Author: {auth}')
    
    #----rating-----------------------------------------------
    
    rating = soup.find_all('div', class_= 'RatingStatistics__rating')[0].text
    print(f'Overall Rating: {rating}')
    
    #----genre------------------------------------------------
    
    genre = soup.find('a', {'class': 'Button Button--tag-inline Button--small'}).get_text()
    print(f'Genre: {genre}')
    
    #----rating and review counts------------------------------
    
    stats = soup.find_all('div', class_ = "RatingStatistics__meta")
    
    rating_span = soup.find('span', {'data-testid': 'ratingsCount'})
    review_span = soup.find('span', {'data-testid': 'reviewsCount'})
    
    rating_text = rating_span.text.strip()
    review_text = review_span.text.strip()
    
    rating_text = re.sub('[^0-9]', '', rating_text)
    review_text = re.sub('[^0-9]', '', review_text)
    print(f'Number of ratings: {rating_text}, Number of reviews: {review_text}')
         
    #----summary------------------------------
    
    summary = soup.find_all('span', class_ = "Formatted")[0].text
           
    #----# of pages------------------------------
    
    page = soup.find('p').text
    page = re.sub('[^0-9]', '', page)
    print(f'Number of pages: {page}')
          
    #--------year--------------------------
    
    year = (soup.find_all('p')[1].text)
    year = year[-4:]
    print(f'Year published: {year}')
    
    #--------reviews--------------------------
        
    reviews = soup.find_all('section',class_ = "ReviewText__content")
          
    ls_reviews = []
    for i in range (0,len (reviews)):
        rev = soup.find_all('section', class_ = "ReviewText__content")[i].text
        ls_reviews.append(rev)
    
    print('\n')
    
    time.sleep(1)
    
    #new_row = pd.Series({'title': title, 'summary': summary, 'year_published': year, 'author': auth, 'review_count': review_text, 
    #                     'number_of_ratings': rating_text, 'length': page, 'genre':genre, 'rating':rating, 'reviews':ls_reviews})
    
    #pd.concat([collect, new_row.to_frame().T], ignore_index=True)
    
    counter = collect.loc[counter, ['title','summary','year_published','author','review_count',
                                  'number_of_ratings','length','genre','rating','reviews']] = title, summary, year, auth, review_text, rating_text, page, genre, rating, reviews
    counter = counter + 1 
```
