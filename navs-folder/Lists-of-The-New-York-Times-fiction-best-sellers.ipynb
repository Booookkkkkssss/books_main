# creating golden list of all books on NY best sellers list

```python
import pandas as pd
import numpy as np 

import time
from requests import get
```

```python
base_url = 'https://en.wikipedia.org/wiki/The_New_York_Times_Fiction_Best_Sellers_of_'
```

# 1931 - 2020`

```python
all_books = pd.DataFrame()
for i in range(1931, 2021):
    url = f'{base_url}{i}'
    tables = pd.read_html(url)
    if i == 1946:
        df = tables[1]
    else:
        df = tables[0]
    if 'Publisher' in df.columns:
        df.drop(columns=['Publisher'], inplace=True)
    if i >= 2021:
        df.rename(columns={'Date': 'Issue Date'}, inplace=True)
    else:
        df.rename(columns={'Date': 'Date'}, inplace=True)
    df['Date'] = df['Date'].apply(lambda x: str(x) + f' {i}')
    all_books = pd.concat([all_books, df], ignore_index=True)
    time.sleep(1)
    print(i)
    print('\n')
```
